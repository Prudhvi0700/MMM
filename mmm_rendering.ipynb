```python
# Create a patch for chumpy/ch.py and chumpy/linalg.py
import inspect
import os

def fix_chumpy_files():
    # Fix ch.py file
    ch_file_path = "/usr/local/lib/python3.11/dist-packages/chumpy/ch.py"
    if os.path.exists(ch_file_path):
        with open(ch_file_path, 'r') as file:
            content = file.read()

        # Replace getargspec with getfullargspec
        content = content.replace('inspect.getargspec', 'inspect.getfullargspec')

        with open(ch_file_path, 'w') as file:
            file.write(content)
        print(f"Fixed {ch_file_path}")

    # Fix linalg.py file
    linalg_file_path = "/usr/local/lib/python3.11/dist-packages/chumpy/linalg.py"
    if os.path.exists(linalg_file_path):
        with open(linalg_file_path, 'r') as file:
            content = file.read()

        # Replace getargspec with getfullargspec
        content = content.replace('inspect.getargspec', 'inspect.getfullargspec')

        with open(linalg_file_path, 'w') as file:
            file.write(content)
        print(f"Fixed {linalg_file_path}")

    # Need to also fix other chumpy files that might use getargspec
    chumpy_dir = "/usr/local/lib/python3.11/dist-packages/chumpy"
    for filename in os.listdir(chumpy_dir):
        if filename.endswith('.py') and filename not in ['ch.py', 'linalg.py']:
            file_path = os.path.join(chumpy_dir, filename)
            with open(file_path, 'r') as file:
                content = file.read()

            if 'inspect.getargspec' in content:
                content = content.replace('inspect.getargspec', 'inspect.getfullargspec')
                with open(file_path, 'w') as file:
                    file.write(content)
                print(f"Fixed {file_path}")

# Run the fix
fix_chumpy_files()
```

```python
%%writefile /content/MMM-Demo/render_final.py
from models.rotation2xyz import Rotation2xyz
import numpy as np
from trimesh import Trimesh
import os
import torch
from visualize.simplify_loc2rot import joints2smpl
import pyrender
import imageio
from shapely import geometry
import trimesh
from pyrender.constants import RenderFlags
import moviepy.editor as mp
import argparse

# Configure GPU rendering if available
import platform
system_type = platform.system()

# Check if CUDA is available
CUDA_AVAILABLE = torch.cuda.is_available()
GPU_DEVICE = torch.device('cuda:0' if CUDA_AVAILABLE else 'cpu')

# Setup appropriate OpenGL platform
# if system_type == 'Linux':
#     # Try EGL first (works better on systems with GPU)
#     if CUDA_AVAILABLE:
#         os.environ['PYOPENGL_PLATFORM'] = 'egl'
#     else:
#         os.environ['PYOPENGL_PLATFORM'] = 'osmesa'  # Software rendering
# else:
#     # For other systems (Windows/Mac) use default
#     pass  # Default OpenGL platform

print(f"Using device: {GPU_DEVICE}, OpenGL platform: {os.environ.get('PYOPENGL_PLATFORM', 'default')}")

class WeakPerspectiveCamera(pyrender.Camera):
    def __init__(self,
                 scale,
                 translation,
                 znear=pyrender.camera.DEFAULT_Z_NEAR,
                 zfar=None,
                 name=None):
        super(WeakPerspectiveCamera, self).__init__(
            znear=znear,
            zfar=zfar,
            name=name,
        )
        self.scale = scale
        self.translation = translation

    def get_projection_matrix(self, width=None, height=None):
        P = np.eye(4)
        P[0, 0] = self.scale[0]
        P[1, 1] = self.scale[1]
        P[0, 3] = self.translation[0] * self.scale[0]
        P[1, 3] = -self.translation[1] * self.scale[1]
        P[2, 2] = -1
        return P

def render(motions, outdir='test_vis', device_id=0, name=None, pred=True):
    frames, njoints, nfeats = motions.shape
    MINS = motions.min(axis=0).min(axis=0)
    MAXS = motions.max(axis=0).max(axis=0)

    height_offset = MINS[1]
    motions[:, :, 1] -= height_offset

    # Set the appropriate device - use CUDA if available and requested
    device = torch.device(f'cuda:{device_id}' if CUDA_AVAILABLE and device_id >= 0 else 'cpu')
    print(f"Using device: {device} for SMPL calculations")

    j2s = joints2smpl(num_frames=frames, device_id=device_id, cuda=CUDA_AVAILABLE and device_id >= 0)
    rot2xyz = Rotation2xyz(device=device)
    faces = rot2xyz.smpl_model.faces

    filepath = os.path.join(outdir, f'{name}.pt')
    if not os.path.exists(filepath):
        print(f'Running SMPLify, it may take a few minutes.')
        motion_tensor, opt_dict = j2s.joint2smpl(motions)

        # Move data to selected device
        motion_tensor_device = torch.tensor(motion_tensor, device=device).clone()

        vertices = rot2xyz(motion_tensor_device, mask=None,
                          pose_rep='rot6d', translation=True, glob=True,
                          jointstype='vertices', vertstrans=True)

        # Save to disk
        torch.save(vertices.cpu(), filepath)
    else:
        print(f'Loading pre-calculated vertices from {filepath}')
        # Load vertices and move to selected device if needed
        vertices = torch.load(filepath, map_location=device)

    if not os.path.exists(outdir):
        os.makedirs(outdir)

    frames = vertices.shape[3]  # shape: 1, nb_frames, 3, nb_joints
    print(f"Rendering {frames} frames of shape {vertices.shape}")

    # Calculate limits for the scene
    MINS = torch.min(torch.min(vertices[0], axis=0)[0], axis=1)[0]
    MAXS = torch.max(torch.max(vertices[0], axis=0)[0], axis=1)[0]

    minx = MINS[0] - 0.5
    maxx = MAXS[0] + 0.5
    minz = MINS[2] - 0.5
    maxz = MAXS[2] + 0.5

    # Create polygon for the floor
    polygon = geometry.Polygon([[minx.cpu().numpy(), minz.cpu().numpy()],
                               [minx.cpu().numpy(), maxz.cpu().numpy()],
                               [maxx.cpu().numpy(), maxz.cpu().numpy()],
                               [maxx.cpu().numpy(), minz.cpu().numpy()]])
    polygon_mesh = trimesh.creation.extrude_polygon(polygon, 1e-5)

    vid = []
    render_frames = frames
    for i in range(render_frames):
        if i % 10 == 0:
            print(f"Rendering frame {i}/{render_frames}")

        # Get vertices for current frame and convert to numpy
        frame_verts = vertices[0, :, :, i].squeeze().cpu().numpy()
        mesh = Trimesh(vertices=frame_verts, faces=faces)

        base_color = (0.11, 0.53, 0.8, 0.5)
        material = pyrender.MetallicRoughnessMaterial(
            metallicFactor=0.7,
            alphaMode='OPAQUE',
            baseColorFactor=base_color
        )

        mesh = pyrender.Mesh.from_trimesh(mesh, material=material)

        polygon_mesh.visual.face_colors = [0, 0, 0, 0.21]
        polygon_render = pyrender.Mesh.from_trimesh(polygon_mesh, smooth=False)

        bg_color = [1, 1, 1, 0.8]
        scene = pyrender.Scene(bg_color=bg_color, ambient_light=(0.4, 0.4, 0.4))

        camera = pyrender.PerspectiveCamera(yfov=(np.pi / 3.0))
        light = pyrender.DirectionalLight(color=[1,1,1], intensity=300)

        scene.add(mesh)

        c = np.pi / 2
        scene.add(polygon_render, pose=np.array([
            [1, 0, 0, 0],
            [0, np.cos(c), -np.sin(c), MINS[1].cpu().numpy()],
            [0, np.sin(c), np.cos(c), 0],
            [0, 0, 0, 1]
        ]))

        # Add lights from different angles
        light_pose = np.eye(4)
        light_pose[:3, 3] = [0, -1, 1]
        scene.add(light, pose=light_pose.copy())

        light_pose[:3, 3] = [0, 1, 1]
        scene.add(light, pose=light_pose.copy())

        light_pose[:3, 3] = [1, 1, 2]
        scene.add(light, pose=light_pose.copy())

        # Position camera
        c = -np.pi / 6
        minx_val = minx.cpu().numpy()
        maxx_val = maxx.cpu().numpy()
        minz_val = minz.cpu().numpy()
        mins_y_val = MINS[1].cpu().numpy()

        scene.add(camera, pose=[
            [1, 0, 0, (minx_val + maxx_val) / 2],
            [0, np.cos(c), -np.sin(c), 1.5],
            [0, np.sin(c), np.cos(c), max(4, minz_val + (1.5 - mins_y_val) * 2, (maxx_val - minx_val))],
            [0, 0, 0, 1]
        ])

        # Render scene with GPU support if available
        r = pyrender.OffscreenRenderer(960, 960)
        color, _ = r.render(scene, flags=RenderFlags.RGBA)
        vid.append(color)
        r.delete()

    # Save as GIF and MP4
    out = np.stack(vid, axis=0)
    gif_path = os.path.join(outdir, f'{name}.gif')
    mp4_path = os.path.join(outdir, f'{name}.mp4')

    print(f"Saving GIF to {gif_path}")
    imageio.mimwrite(gif_path, out, fps=20)

    print(f"Converting to MP4: {mp4_path}")
    out_video = mp.VideoFileClip(gif_path)
    out_video.write_videofile(mp4_path, codec='libx264', fps=20)
    print("Rendering complete!")

    return mp4_path

if __name__ == "__main__":
    parser = argparse.ArgumentParser(description="Render 3D motion from a numpy file.")
    parser.add_argument("filepath", type=str, help="Path to the numpy file containing the motion data.")
    parser.add_argument("--outdir", type=str, default="./output", help="Output directory for rendered files.")
    parser.add_argument("--gpu", type=int, default=0, help="GPU device ID to use (-1 for CPU)")
    args = parser.parse_args()

    # Check if GPU should be used
    use_gpu = args.gpu >= 0 and torch.cuda.is_available()
    if args.gpu >= 0 and not torch.cuda.is_available():
        print("Warning: GPU requested but CUDA is not available. Falling back to CPU.")

    device_id = args.gpu if use_gpu else -1

    # Ensure output directory exists
    if not os.path.exists(args.outdir):
        os.makedirs(args.outdir)

    # Load the motion data from the provided file path
    motions = np.load(args.filepath)
    print('Loaded motion data from:', args.filepath, 'Shape:', motions.shape)

    # Extract filename for naming output files
    filename = os.path.basename(args.filepath).replace('.npy', '')

    # Render the motion
    render(motions[0], outdir=args.outdir, device_id=device_id, name=filename)
```

```python
import subprocess
import os
import glob
import base64
from IPython.display import HTML, Video, display as ipython_display

# Function to find the latest file with a given extension in a specified directory
def find_latest_file(base_path, extension):
    list_of_files = glob.glob(f'{base_path}/*.{extension}')
    if not list_of_files:
        return None
    latest_file = max(list_of_files, key=os.path.getctime)
    return latest_file

# Function to run the generate.py script and return paths of generated HTML and NPY files
def generate_html(text_input, length):
    print(f"Generating motion for: '{text_input}' with length {length}...")
    command = [
        "python", "generate.py",
        "--resume-pth", "output/vq/2023-07-19-04-17-17_12_VQVAE_20batchResetNRandom_8192_32/net_last.pth",
        "--resume-trans", "output/t2m/2023-10-10-03-17-01_HML3D_44_crsAtt2lyr_mask0.5-1/net_last.pth",
        "--text", text_input,
        "--length", str(length)
    ]
    try:
        result = subprocess.run(command, check=True, text=True, capture_output=True)
        html_file = find_latest_file('output', 'html')
        npy_file = find_latest_file('output', 'npy')
        print(f"Generated files: HTML: {html_file}, NPY: {npy_file}")
        return html_file, npy_file
    except subprocess.CalledProcessError as e:
        print(f"Error: {e.stderr}")
        return None, None

# Function to run render_final.py script with the generated NPY file
def run_render_final(npy_file_path, use_gpu=True):
    print(f"Rendering skeleton for: {npy_file_path}...")

    # Check if GPU is available
    gpu_available = torch.cuda.is_available()
    gpu_flag = "--gpu" if use_gpu and gpu_available else "--gpu -1"

    if use_gpu:
        if gpu_available:
            print(f"Using GPU: {torch.cuda.get_device_name(0)}")
        else:
            print("GPU requested but not available. Falling back to CPU rendering.")

    # Construct the command with GPU flag if needed
    if use_gpu and gpu_available:
        command = ["python", "render_final.py", npy_file_path, "--gpu", "0"]
    else:
        command = ["python", "render_final.py", npy_file_path, "--gpu", "-1"]

    try:
        result = subprocess.run(command, check=True, text=True, capture_output=True)
        vid_file_path = find_latest_file('output', 'mp4')
        print(f"Generated video: {vid_file_path}")
        return vid_file_path
    except subprocess.CalledProcessError as e:
        print(f"Error in render_final.py: {e.stderr}")
        return None

# Function to display HTML content in the notebook
def display_html(html_file_path):
    if html_file_path and os.path.exists(html_file_path):
        with open(html_file_path, 'r') as file:
            html_content = file.read()
        return HTML(html_content)
    return None

# Main function to run the pipeline - generate only, no rendering
def generate_motion_from_text(text_input, length):
    # Generate HTML and NPY files without trying to render video
    html_file_path, npy_file_path = generate_html(text_input, length)

    if not html_file_path or not npy_file_path:
        print("Error generating files. Please try again.")
        return None

    # Display the HTML output
    html_output = display_html(html_file_path)
    return html_output, npy_file_path
```

```python
%%writefile streamlit_app.py
import streamlit as st
import subprocess
import os
import glob
import base64
import time
import torch
from pathlib import Path
import numpy as np
from PIL import Image
import tempfile

# Set page configuration
st.set_page_config(
    page_title="Text to Motion Generator",
    page_icon="🏃",
    layout="wide",
    initial_sidebar_state="expanded"
)

# Custom CSS to make the app more attractive
st.markdown("""
<style>
    .main-header {
        font-size: 2.5rem;
        font-weight: 700;
        color: #1E88E5;
        text-align: center;
        margin-bottom: 1rem;
    }
    .sub-header {
        font-size: 1.5rem;
        font-weight: 500;
        color: #424242;
        margin-bottom: 1rem;
    }
    .stButton > button {
        background-color: #1E88E5;
        color: white;
        font-weight: bold;
        padding: 0.5rem 1rem;
        border-radius: 5px;
        border: none;
        transition: all 0.3s;
    }
    .stButton > button:hover {
        background-color: #1565C0;
        transform: translateY(-2px);
        box-shadow: 0 4px 6px rgba(0,0,0,0.1);
    }
    .render-button > button {
        background-color: #43A047;
    }
    .render-button > button:hover {
        background-color: #2E7D32;
    }
    .output-container {
        background-color: #f5f5f5;
        padding: 20px;
        border-radius: 10px;
        box-shadow: 0 4px 6px rgba(0,0,0,0.1);
        margin-top: 20px;
    }
    .stSpinner > div {
        border-color: #1E88E5 transparent transparent transparent !important;
    }
</style>
""", unsafe_allow_html=True)

# Initialize session state variables
if 'html_file_path' not in st.session_state:
    st.session_state['html_file_path'] = None
if 'npy_file_path' not in st.session_state:
    st.session_state['npy_file_path'] = None
if 'video_file_path' not in st.session_state:
    st.session_state['video_file_path'] = None
if 'generation_completed' not in st.session_state:
    st.session_state['generation_completed'] = False
if 'render_completed' not in st.session_state:
    st.session_state['render_completed'] = False

# Function to find the latest file with a given extension in a specified directory
def find_latest_file(base_path, extension):
    list_of_files = glob.glob(f'{base_path}/*.{extension}')
    if not list_of_files:
        return None
    latest_file = max(list_of_files, key=os.path.getctime)
    return latest_file

# Function to run the generate.py script and return paths of generated HTML and NPY files
def generate_html(text_input, length):
    command = [
        "python", "generate.py",
        "--resume-pth", "output/vq/2023-07-19-04-17-17_12_VQVAE_20batchResetNRandom_8192_32/net_last.pth",
        "--resume-trans", "output/t2m/2023-10-10-03-17-01_HML3D_44_crsAtt2lyr_mask0.5-1/net_last.pth",
        "--text", text_input,
        "--length", str(length)
    ]
    try:
        result = subprocess.run(command, check=True, text=True, capture_output=True)
        html_file = find_latest_file('output', 'html')
        npy_file = find_latest_file('output', 'npy')
        return html_file, npy_file
    except subprocess.CalledProcessError as e:
        st.error(f"Error: {e.stderr}")
        return None, None

# Function to run render_final.py script with the generated NPY file
def run_render_final(npy_file_path, use_gpu=True):
    # Check if GPU is available
    gpu_available = torch.cuda.is_available()

    # Construct the command with GPU flag if needed
    if use_gpu and gpu_available:
        command = ["python", "render_final.py", npy_file_path, "--gpu", "0"]
    else:
        command = ["python", "render_final.py", npy_file_path, "--gpu", "-1"]

    try:
        result = subprocess.run(command, check=True, text=True, capture_output=True)
        vid_file_path = find_latest_file('output', 'mp4')
        return vid_file_path
    except subprocess.CalledProcessError as e:
        st.error(f"Error in render_final.py: {e.stderr}")
        return None

# Function to display HTML content in Streamlit
def display_html(html_file_path):
    if html_file_path and os.path.exists(html_file_path):
        with open(html_file_path, 'r') as file:
            html_content = file.read()
        st.components.v1.html(html_content, height=500)
        return True
    return False

# Function to get base64 encoded video
def get_video_base64(video_file):
    if not os.path.exists(video_file):
        return None

    with open(video_file, 'rb') as f:
        data = f.read()
        base64_video = base64.b64encode(data).decode()
        return base64_video

# Generate skeleton animation
def generate_motion():
    with st.spinner("Generating skeleton animation... This may take a moment"):
        html_file_path, npy_file_path = generate_html(
            st.session_state.text_input,
            st.session_state.length
        )

        if html_file_path and npy_file_path:
            st.session_state['html_file_path'] = html_file_path
            st.session_state['npy_file_path'] = npy_file_path
            st.session_state['generation_completed'] = True
            st.success("Skeleton animation generated successfully!")
        else:
            st.error("Failed to generate skeleton animation. Please try again.")

# Render video from skeleton
def render_video():
    with st.spinner("Rendering video... This may take a while"):
        video_file_path = run_render_final(
            st.session_state.npy_file_path,
            use_gpu=st.session_state.use_gpu
        )

        if video_file_path:
            st.session_state['video_file_path'] = video_file_path
            st.session_state['render_completed'] = True
            st.success("Video rendered successfully!")
        else:
            st.error("Failed to render video. Please try again.")

# Main App UI
st.markdown('<h1 class="main-header">🏃 Text to Motion Generator</h1>', unsafe_allow_html=True)
st.markdown('<p class="sub-header">Transform your text descriptions into skeleton animations and videos</p>', unsafe_allow_html=True)

# Sidebar for instructions and options
with st.sidebar:
    st.markdown("## How to use")
    st.markdown("""
    1. Enter your motion description
    2. Set the animation length
    3. Click 'Generate Skeleton' button
    4. After the skeleton is generated, click 'Render Video'
    """)

    st.markdown("## About")
    st.markdown("""
    This app converts text descriptions into human motion animations using a neural network model.

    The process has two steps:
    1. Generate a skeleton animation
    2. Render the animation into a video
    """)

    st.markdown("## Rendering Options")
    st.session_state.use_gpu = st.checkbox("Use GPU for rendering (if available)", value=True)

    # GPU availability check
    if st.session_state.use_gpu:
        if torch.cuda.is_available():
            st.success(f"GPU available: {torch.cuda.get_device_name(0)}")
        else:
            st.warning("No GPU detected. Will use CPU for rendering.")

# Input section
col1, col2 = st.columns([3, 1])

with col1:
    st.text_input(
        "Enter motion description:",
        key="text_input",
        value="A person walks forward then turns completely around and does a cartwheel",
        help="Describe the motion you want to generate"
    )

with col2:
    st.number_input(
        "Animation Length:",
        min_value=60,
        max_value=300,
        value=196,
        step=4,
        key="length",
        help="Longer animations require more processing time"
    )

# Generate button
generate_col, _ = st.columns([1, 3])
with generate_col:
    if st.button("Generate Skeleton", use_container_width=True):
        generate_motion()

# Display the skeleton animation if generated
if st.session_state.generation_completed:
    # st.markdown('<div class="output-container">', unsafe_allow_html=True)
    st.markdown('<h2 class="sub-header">Skeleton Animation</h2>', unsafe_allow_html=True)

    if display_html(st.session_state.html_file_path):
        # Render button
        render_col, _ = st.columns([1, 3])
        with render_col:
            st.markdown('<div class="render-button">', unsafe_allow_html=True)
            if st.button("Render Video", use_container_width=True):
                render_video()
            st.markdown('</div>', unsafe_allow_html=True)

    st.markdown('</div>', unsafe_allow_html=True)

# Display the rendered video if completed
if st.session_state.render_completed:
    # st.markdown('<div class="output-container">', unsafe_allow_html=True)
    st.markdown('<h2 class="sub-header">Rendered Video</h2>', unsafe_allow_html=True)

    video_file = st.session_state.video_file_path
    if video_file and os.path.exists(video_file):
        video_bytes = open(video_file, 'rb').read()
        st.video(video_bytes)

        # Download button for the video
        st.download_button(
            label="Download Video",
            data=video_bytes,
            file_name=os.path.basename(video_file),
            mime="video/mp4"
        )
    else:
        st.error("Video file not found.")

    st.markdown('</div>', unsafe_allow_html=True)

# Footer
st.markdown("""
---
<p style="text-align: center; color: #666;">
    Created with Streamlit • Text-to-Motion Generator
</p>
""", unsafe_allow_html=True)
```
